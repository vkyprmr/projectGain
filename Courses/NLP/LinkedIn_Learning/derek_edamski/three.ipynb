```python
ps = nltk.PorterStemmer()
wn = nltk.WordNetLemmatizer()
```

```python
df = pd.read_csv(data_dir / 'SMSSpamCollection.tsv', sep='\t', header=None)
df.columns = ['label', 'text']
```

```python
def clean_text(text):
    text = ''.join([c for c in text if c not in string.punctuation])
    text = re.split('\W+', text.lower())
    stopwords = nltk.corpus.stopwords.words('english')
    text = [c for c in text if c not in stopwords]
    return [wn.lemmatize(w) for w in text]
```

```python
# Count vectorization
```

```python
count_vetorizer = CountVectorizer(analyzer=clean_text)
x_counts = count_vetorizer.fit_transform(df['text'])
print(x_counts.shape)
print(count_vetorizer.get_feature_names_out())
```

```python
def clean_text_n_grams(text):
    text = ''.join([c for c in text if c not in string.punctuation])
    text = re.split('\W+', text.lower())
    stopwords = nltk.corpus.stopwords.words('english')
    text = " ".join([c for c in text if c not in stopwords])
    return text

df['cleaned_text'] = df['text'].apply(lambda x: clean_text_n_grams(x))
df.head()
```

```python
count_vect = CountVectorizer(ngram_range=(2,3))
x_counts_n_grams = count_vect.fit_transform(df.cleaned_text)
print(x_counts_n_grams.shape)
print(count_vect.get_feature_names_out())
```

```python
nrange_df = pd.DataFrame(x_counts_n_grams.toarray(), columns=count_vect.get_feature_names_out())
```

```python
# TF-IDF
"""
wi_j = tfi_j * log(N/dfi)
tfi_j = number of times i occurs in j divided by total number of terms in j
dfi = number of documents containing i
N = total number of documents
"""
```

```python
def clean_text(text):
    text = ''.join([c for c in text if c not in string.punctuation])
    text = re.split('\W+', text.lower())
    stopwords = nltk.corpus.stopwords.words('english')
    text = [c for c in text if c not in stopwords]
    return [wn.lemmatize(w) for w in text]
```

```python
tfidf = TfidfVectorizer(analyzer=clean_text)
x_tfidf = tfidf.fit_transform(df.text)
print(x_tfidf.shape)
print(tfidf.get_feature_names_out())
```

```python
tfidf_df = pd.DataFrame(x_tfidf.toarray(), columns=tfidf.get_feature_names_out())
```
